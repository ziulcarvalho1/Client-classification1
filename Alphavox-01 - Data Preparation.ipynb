```python
## Este programa tem como objetivo completar o campo sexo do registro quando o mesmo não estiver 
## previamente preenchido. Isto é feito baseado no primeiro nome do indivíduo. O programa também atualiza 
## a base histórica identificando quantos indivíduos do sexo masculino e quantos do sexo feminino foram 
## identificados com aquele nome em particular. Quando nos deparamos com o nome sem a identificação de sexo 
## ele preenche o campo com o sexo com maior quantidade de amostras  associado ao nome. Utiliza o arquivo 
## nome-y.csv como depositório dos nomes.
```

```python
from datetime import datetime
count=0
count2=1000
for i in range(len(df1)):
    if i==count2:
        now = datetime.now()
        timestamp = datetime.timestamp(now)
        dt_object = datetime.fromtimestamp(timestamp)
        print(dt_object,count2,i)
        print('Preenchi',count)
        count2=count2+1000
    if df1['SEXO'][i]!="M" and df1['SEXO'][i]!="F":
        df1['SEXO'][i]=" "
    achei=0   
    tnome = df1['NOME'][i].strip()
    tnome1=tnome.split()
    tnome2=tnome1[0]
    tsexo1 =df1['SEXO'][i]
    #print(tnome2)
    for i1 in range(len(df2)):
        if df2['name'][i1]==tnome2:
            achei = 1
            #print('Atualizei')
            if df1['SEXO'][i]=="M":
                df2['SEXO_M'][i1]=df2['SEXO_M'][i1]+1
            if df1['SEXO'][i]=="F":
                df2['SEXO_F'][i1]=df2['SEXO_F'][i1]+1
            if df1['SEXO'][i]==" ":
                #print('preenchi',count)
                count=count+1
                if df2['SEXO_F'][i1]>df2['SEXO_M'][i1]:
                    tsexo="F"
                else:
                    tsexo="M"
                df1['SEXO'][i]=tsexo
            break
    if achei==0:
        #print('Adicionei')
        tsexo_M=0
        tsexo_F=0
        if tsexo1=="M":
            tsexo_M=1
            tsexo_F=0
        if tsexo1=="F":
            tsexo_F=1
            tsexo_M=0
        df2 = df2.append({'name':tnome2,'SEXO_M':tsexo_M,'SEXO_F':tsexo_F}, ignore_index=True)  
        df2 = df2.reindex()
    
```

```python
#Este codigo calcula a idade do individuo baseado em sua data de nascimento e associa a uma 
# faixa de idade de 15 a 95 anos em faixas de 5 anos.
```

```python
from datetime import datetime
now = datetime.now()
timestamp = datetime.timestamp(now)
dt_object = datetime.fromtimestamp(timestamp)
dt_object1 = dt_object.strftime("%Y, %m %d %H")
print(dt_object1)
count=0
count2=1000
df1['idade']=0
df1['idade_range']=""
for i in range(len(df1)):
    if i==count2:
        now = datetime.now()
        timestamp = datetime.timestamp(now)
        dt_object = datetime.fromtimestamp(timestamp)
        print(dt_object,count2,i)
        print('Preenchi',count)
        count2=count2+1000
    tdata=df1['DATA_NASCIMENTO'][i]
    tano=tdata[6:]
    tanoatual=dt_object1[0:4]
    #print(tano,tanoatual)
    tidade=int(tanoatual)-int(tano)
    df1['idade'][i]=tidade
    y = 1
    for x in range(15,95,5):
        if tidade>=x and tidade<x+5:
            if y<10:
                tfaixa = "FAIXA-0"+str(y)
            else:
                tfaixa = "FAIXA-"+str(y) 
        y=y+1
    df1['idade_range'][i]=tfaixa
```
