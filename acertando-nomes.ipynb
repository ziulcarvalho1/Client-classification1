```python
df3 = pd.DataFrame(columns=['name','SEXO_M','SEXO_F'])
```

```python

count2=1000
for i in range(len(df2)):
    count=0
    
    if i==count2:
        print(count,count2,i)
        count2=count2+1000
        count=count+1
    tname=df2['name'][i]
    tsexo=df2['sexo'][i]
    tquant=df2['quant'][i]
    achei = 0
    for i1 in range(len(df3)):
        if tname==df3['name'][i1]:
            achei=1
            if tsexo=="M":
                df3['SEXO_M'][i1]=df3['SEXO_M'][i1]+tquant
            if tsexo=="F":
                df3['SEXO_F'][i1]=df3['SEXO_F'][i1]+tquant
            break
    if achei==0:
        if tsexo=="M":
            tsexo_M=tquant
            tsexo_F=0
        if tsexo=="F":
            tsexo_F=tquant
            tsexo_M=0
        df3 = df3.append({'name':tname,'SEXO_M':tsexo_M,'SEXO_F':tsexo_F}, ignore_index=True)  
        df3 = df3.reindex()
```
